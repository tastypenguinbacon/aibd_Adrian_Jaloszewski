Zdefiniuj poniższą funkcję i sporządź jej wykres dla argumentów z danego przedziału:
$f(x) = x^2 + 5$
1. $x > -1\ \mathrm{oraz}\ x < 1$
2. $x > -6\ \mathrm{oraz}\ x < 6$
3. $x > 0\ \mathrm{oraz}\ x < 6$

Poniższa komórka zawiera importowane biblioteki `numpy` w celu definicji wektora i obliczeń wektorowych oraz `matplotlib.pyplot` w celu wizualizacji wyników w postaci wykresu.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Definicja granic przedziałów otwartych, w których zdefiniowana jest funkcja.

```python
limits = [
    (-1, 1),
    (-6, 6),
    (0, 6)
]
```

Wydzielony kod do rysowania wykresu z przedziałami otwartymi:

```python
def plot_graph(xs, ys):
    plt.figure()
    plt.grid(True)
    plt.ylabel(r'$f(x) = x^2 + 5$')
    plt.xlabel('$x$')
    # Przedział otwarty
    plt.plot(xs[1:-1], ys[1:-1], label=f'${xs[0]} < x < {xs[-1]}$')
    # Puste koła na brzegach przedziału otwartego
    plt.scatter([xs[0], xs[-1]], [ys[0], ys[-1]], facecolors='none', edgecolors='C0', s=80, linewidths=3)
    plt.legend()
```

Obliczenie wartości funkcji $f(x) = x^2 +5$ oraz wywołanie funkcji do rysowania wykresu.

```python
for min_x, max_x in limits:
    xs = np.linspace(min_x, max_x, 200)
    ys = xs ** 2 + 5
    plot_graph(xs, ys)
```
